```python
import pandas as pd
import matplotlib.pyplot as plt
print("Trần Lương Hòa, MSSV: 2174802010453")
```

```python
records = [('Alice','25','50000'), 
           ('Bob', '30', '60000'), 
           ('Charlie', '35',"70000"), 
           ('David', '28',"55000" ),
           ('Eva', '22',"52000"), 
           ('Frank', '45',"80000" ),
           ('Grace', '34',"72000"), 
           ('Hannah', '31',"68000" ),
           ('Ivan', '27',"61000"), 
           ('Jack', '29',"59000" ),
           ('Kelly', '33',"63000" ),
           ('Liam', '40',"77000"), 
           ('Mona', '26',"53000" ),
           ('Nina', '32',"66000" ),
           ('Oscar', '36',"75000" )]
           
           
# Khởi tạo DataFrame
df = pd.DataFrame(data = records, columns = ['Name', 'Age', 'Salary'])
df
```

```python
#thay thế các giá trị của cột "Age" lớn hơn 30 bằng giá trị trung bình của cột đó
# Tính giá trị trung bình của cột 'Age'trước
age_mean = df['Age'].mean()

# Thay thế các giá trị của cột 'Age' lớn hơn 30 bằng giá trị trung bình
df.loc[df['Age'] > 30, 'Age'] = age_mean

print("DataFrame sau khi thay thế các giá trị của cột 'Age' lớn hơn 30 bằng giá trị trung bình:")
print(df)
```

```python
#chuẩn hóa (normalize) cột 'Age' về khoảng giá trị 0 đến 1
df['age_mean']=(df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())
df
```

```python
# Create cột mới là 'Age_group' phân loại tuổi thành 'Young','Middle-aged' và 'Old' dựa trên giá trị của cột 'Age', young là dưới 30 tuổi , middle_Aged là từ 30 tuổi dưới 60 tuổi , old là trên 60 tuổi
def classify_age(age):
    if age < 30:
        return 'Young'
    elif 30 <= age < 60:
        return 'Middle-aged'
    else:
        return 'Old'

# Tạo cột mới 'Age_group' dựa trên giá trị của cột 'Age'
df['Age_group'] = df['Age'].apply(classify_age)

print("DataFrame sau khi phân loại độ tuổi:")
print(df)
```

```python
#tính toán tỷ lệ phần trăm thay đổi (precentage change) của cột 'Salary'
df['Salary_pct_change'] = df['Salary'].pct_change() * 100
df
```

```python
#tìm các giá trị trùng lặp trong DataFarm dựa trên cột 'Name' và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên 
df.drop_duplicates(subset=['Name'], keep='first', inplace=True)
df
```
