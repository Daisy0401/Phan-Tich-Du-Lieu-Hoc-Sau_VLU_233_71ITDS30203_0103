```python
def xep_loai(tbm):
    if tbm < 5.0:
        return 'Y'
    elif 5.0 <= tbm < 6.5:
        return 'TB'
    elif 6.5 <= tbm < 8.0:
        return 'K'
    elif 8.0 <= tbm < 9.0:
        return 'G'
    else:
        return 'XS'
```

```python
# Chuyển đổi điểm từ thang điểm 10 sang thang điểm 4 bằng phương pháp Min-Max Normalization
def min_max_normalization(tbm, min_old, max_old, min_new, max_new):
    return ((tbm - min_old) / (max_old - min_old)) * (max_new - min_new) + min_new
```

```python
min_old, max_old = 0, 10  # Thang điểm 10 của Việt Nam
min_new, max_new = 0, 4   # Thang điểm 4 của Mỹ
```

```python
df['US_TBM1'] = df['TBM1'].apply(min_max_normalization, args=(min_old, max_old, min_new, max_new))
df['US_TBM2'] = df['TBM2'].apply(min_max_normalization, args=(min_old, max_old, min_new, max_new))
df['US_TBM3'] = df['TBM3'].apply(min_max_normalization, args=(min_old, max_old, min_new, max_new))
```

```python
def xet_tuyen(DH1, DH2, DH3, KT):
    if KT in ['A', 'A1']:
        score = (DH1*2 + DH2 + DH3) / 4
    elif KT == 'B':
        score = (DH1 + DH2*2 + DH3) / 4
    else:
        score = (DH1 + DH2 + DH3) / 3
    
    return 1 if score >= 5.0 else 0
```

```python
df['KQXT'] = df.apply(lambda row: xet_tuyen(row['DH1'], row['DH2'], row['DH3'], row['KT']), axis=1)
```
