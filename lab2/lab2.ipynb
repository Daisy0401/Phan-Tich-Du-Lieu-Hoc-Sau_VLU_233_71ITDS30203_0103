```python
# Tạo pivot-table để thống kê các giá trị của DH1 theo KT
pivot_kt = df.pivot_table(values='DH1', index='KT', 
                          aggfunc={
                              'DH1': ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 
                                      lambda x: x.quantile(0.25), 
                                      lambda x: x.quantile(0.5), 
                                      lambda x: x.quantile(0.75)]
                          })
pivot_kt.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q2', 'Q3']
pivot_kt.reset_index(inplace=True)
pivot_kt
```

```python
# Tạo pivo-table để thống kê các giá trị của DH1 theo KT và KV
pivot_kt_kv=df.pivot_table(values="DH1", index=['KT','KV'],
                           aggfunc={'DH1':['count','sum', 'mean', 'median', 'min', 'max', 'std',
                                           lambda x: x.quantile(0.25),
                                           lambda x: x.quantile(0.5),
                                           lambda x: x.quantile(0.75)]
                           })
pivot_kt_kv.columns = ['count','sum', 'mean', 'median', 'min', 'max', 'std', 'Q1','Q2','Q3']
pivot_kt_kv.reset_index(inplace=True)
pivot_kt_kv
```

```python
# Tạo pivo-table để thống kê các giá trị của DH1 theo KT và KV và DT
pivot_kt_kv_dt=df.pivot_table(values="DH1", index=['KT','KV'],
                           aggfunc={'DH1':['count','sum', 'mean', 'median', 'min', 'max', 'std',
                                           lambda x: x.quantile(0.25),
                                           lambda x: x.quantile(0.5),
                                           lambda x: x.quantile(0.75)]
                           })
pivot_kt_kv_dt.columns = ['count','sum', 'mean', 'median', 'min', 'max', 'std', 'Q1','Q2','Q3']
pivot_kt_kv_dt.reset_index(inplace=True)
pivot_kt_kv_dt
```

```python
# tạo hàm để trình bày dữ liệu trước
def plot_histogram(data, title, xlabel):
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=20, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Tần số')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
#Tạo biến phân loại (phanlopt1) cho môn Toán (T1)
data = {
    'T1': [4.5, 6.2, 7.5, 8.1, 3.6, 7.8, 6.9, 5.4, 8.0, 6.1]
}
df = pd.DataFrame(data)

# 1. Vẽ biểu đồ đường Simple cho biến T1
plt.figure(figsize=(10, 6))
plt.plot(df['T1'], marker='o', linestyle='-', color='b')
plt.title('Biểu Đồ Đường Simple cho Biến T1')
plt.xlabel('Index')
plt.ylabel('Giá Trị T1')
plt.grid(True)
plt.show()

# 2. Tạo biến phân loại (phanlopt1) cho môn toán (T1)
def classify_t1(t1):
    if t1 < 5:
        return 'k'
    elif 5 <= t1 < 7:
        return 'tb'
    elif 7 <= t1 < 8:
        return 'k'
    else:
        return 'g'

# Tạo cột mới phanlopt1 dựa trên cột T1
df['phanlopt1'] = df['T1'].apply(classify_t1)

# Kiểm tra kết quả
print(df[['T1', 'phanlopt1']])
```

```python
# Lập bảng tần số cho biến phanlopt1
frequency_table_phanlopt1 = df['phanlopt1'].value_counts().reset_index()
frequency_table_phanlopt1.columns = ['phanlopt1', 'Frequency']
print(frequency_table_phanlopt1)
# Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1
plt.figure(figsize=(10, 6))

for label, df_group in df.groupby('phanlopt1'):
    plt.plot(df_group.index, df_group['T1'], marker='o', linestyle='-', label=label)

plt.title('Biểu Đồ Đường Multiple Line cho Biến T1 Phân Loại bởi Biến phanlopt1')
plt.xlabel('Index')
plt.ylabel('Giá Trị T1')
plt.legend(title='phanlopt1')
plt.grid(True)
plt.show()
```

```python
# Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1
plt.figure(figsize=(10, 6))

for label, df_group in df.groupby('phanlopt1'):
    plt.stem(df_group.index, df_group['T1'], label=label, basefmt=" ")

plt.title('Biểu Đồ Drop-line cho Biến T1 Phân Loại bởi Biến phanlopt1')
plt.xlabel('Index')
plt.ylabel('Giá Trị T1')
plt.legend(title='phanlopt1')
plt.grid(True)
plt.show()
```

```python
# Mô tả độ tập trung và phân tán của dữ liệu T1
t1_mean = df['T1'].mean()
t1_median = df['T1'].median()
t1_std = df['T1'].std()
t1_min = df['T1'].min()
t1_max = df['T1'].max()

print(f"Mean: {t1_mean}")
print(f"Median: {t1_median}")
print(f"Standard Deviation: {t1_std}")
print(f"Min: {t1_min}")
print(f"Max: {t1_max}")

# Vẽ biểu đồ Box-Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['T1'])
plt.title('Biểu Đồ Box-Plot cho Biến T1')
plt.xlabel('Giá Trị T1')
plt.show()

# Vẽ biểu đồ Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['T1'], bins=10, kde=True, color='skyblue')
plt.title('Biểu Đồ Histogram cho Biến T1')
plt.xlabel('Giá Trị T1')
plt.ylabel('Tần suất')
plt.show()

# Kiểm chứng phân phối chuẩn với QQ-Plot
import scipy.stats as stats

plt.figure(figsize=(10, 6))
stats.probplot(df['T1'], dist="norm", plot=plt)
plt.title('QQ-Plot Kiểm Chứng Phân Phối Chuẩn cho Biến T1')
plt.show()
```

```python
# Vẽ biểu đồ Box-Plot theo phân nhóm là giá trị của phanlopt1
plt.figure(figsize=(10, 6))
sns.boxplot(x='phanlopt1', y='T1', data=df)
plt.title('Biểu Đồ Box-Plot cho Biến T1 theo Phân Lớp phanlopt1')
plt.xlabel('phanlopt1')
plt.ylabel('Giá Trị T1')
plt.show()

# Vẽ biểu đồ Histogram theo phân nhóm là giá trị của phanlopt1
plt.figure(figsize=(10, 6))
sns.histplot(df, x='T1', hue='phanlopt1', bins=10, kde=True, multiple='stack')
plt.title('Biểu Đồ Histogram cho Biến T1 theo Phân Lớp phanlopt1')
plt.xlabel('Giá Trị T1')
plt.ylabel('Tần suất')
plt.show()

# Vẽ biểu đồ QQ-Plot theo phân nhóm là giá trị của phanlopt1
plt.figure(figsize=(15, 10))

for i, label in enumerate(df['phanlopt1'].unique()):
    plt.subplot(2, 2, i + 1)
    stats.probplot(df[df['phanlopt1'] == label]['T1'], dist="norm", plot=plt)
    plt.title(f'QQ-Plot cho Biến T1 - phanlopt1: {label}')

plt.tight_layout()
plt.show()
```
